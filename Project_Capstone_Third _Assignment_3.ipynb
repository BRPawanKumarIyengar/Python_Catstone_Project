Here we create a filretred dataframe for Toronto

```python
Test_dataframe = PND.read_csv("Mega_Input_File.CSV")
Toronto_Filtered = Test_dataframe[Test_dataframe.Borough.str.contains("Toronto")]
Toronto_Filtered.to_csv("Toronto_Filtered_CSV.CSV", )
Toronto_Filtered = PND.read_csv("Toronto_Filtered_CSV.CSV")
Toronto_Filtered = Toronto_Filtered.drop('Unnamed: 0',axis=1)
```

```python
map_Toronto = FLM.Map(location = [43.6532, -79.3832], zoom_start=12,tiles ='OpenStreetMap',no_wrap=True)
```

We color code accordoing to the Boroughs .... we give red color to Central Toronto, Green color to Downtown Toronto, Blue Color to East Toronto andblack color to West Toronto. We create a mao that shows all the neighbourhoods and the popup displays the Postal code and Name of Neighbourhoods.

```python
for index, row in Toronto_Filtered.iterrows():
    Latitude_String = row['Latitude']
    Longitude_String = row['Longitude']
    Popup_String = row['Borough']+' Postal Code: ' +row['Postcode']+'  -->  '+row['Neighbourhood']
    if row['Borough'] == 'Central Toronto':
        FLM.CircleMarker(
            [Latitude_String, Longitude_String],
            radius=5,
            popup = Popup_String,
            color ='yellow',
            fill = True,
            fill_color = 'red',
            fill_opacity = 0.7).add_to(map_Toronto)
    elif row['Borough'] == 'Downtown Toronto':
        FLM.CircleMarker(
            [Latitude_String, Longitude_String],
            radius=5,
            popup = Popup_String,
            color ='yellow',
            fill = True,
            fill_color = 'green',
            fill_opacity = 0.7).add_to(map_Toronto)
    elif row['Borough'] == 'East Toronto':
        FLM.CircleMarker(
            [Latitude_String, Longitude_String],
            radius=5,
            popup = Popup_String,
            color ='yellow',
            fill = True,
            fill_color = 'blue',
            fill_opacity = 0.7).add_to(map_Toronto)
    else:
        FLM.CircleMarker(
            [Latitude_String, Longitude_String],
            radius=5,
            popup = Popup_String,
            color ='yellow',
            fill = True,
            fill_color = 'black',
            fill_opacity = 0.7).add_to(map_Toronto)
```

Here we use use Foursquare API to explore all the marked points and get the venue details of All the plavs that we have marked on the map.

```python
for index, row in Toronto_Filtered.iterrows():
    Latitude_String = row['Latitude']
    Longitude_String = row['Longitude']
    LIMIT = 10 # limit of number of venues returned by Foursquare API
    radius = 300 # define radius
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    Latitude_String, 
    Longitude_String, 
    radius, 
    LIMIT)
    results.append(requests.get(url).json())
```
